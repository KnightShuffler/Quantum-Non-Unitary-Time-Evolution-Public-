```python
import numpy as np
def bit_sum(k: int, n: int) -> int:
    bsum = 0
    for i in range(n):
        bsum += k % 2
        k //= 2
    return bsum
n=3
N=2**n
w = np.exp(1j*2*np.pi/N)
```

```python
def f_mat(n: int) -> np.array:
    N = 2**n
    w= np.exp(1j*2*np.pi/N)
    mat = np.zeros((N,N), dtype=complex)
    for I in range(N):
        for J in range(N):
            for K in range(N):
                mat[I,J] += (-1)**(bit_sum(K,n)) * w**((I-J)*K/N)
    mat /= N
    return mat
```

```python
f_mat(1)
```

```python
f2 = f_mat(2)
print(f2)
```
